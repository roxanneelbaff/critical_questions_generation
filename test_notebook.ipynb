```python
import time
from experiments.social_agents import init_and_run_experiment
from experiments.utils import llm_dict

run1 = "4SCORES_VAL_mistral24b_social_n3_Teeo_Sddr"
run2 = "4SCORES_VAL_llama70b_social_n2_Tee_Sddd"
run3 = "4RANKS_VAL_llama8b_social_n3_Teee_Srdd"

for run in [run1, run2, run3]:

    llm_key = run.split("_")[2] + "_"
    name = "TEST_"+ "_".join(run.split("_")[2:])
    agents = int(run.split("_")[4][1])
    t_dict = {"e": "easy_going", "o": "overconfident"}
    traits = str(tuple([t_dict[x] for x in run.split("_")[5][1:] ]))

    s_dict = {"r": "reflect", "d": "debate"}
    strategy = str(tuple([s_dict[x] for x in run.split("_")[6][1:] ]))
    print(llm_key, "llm_key")
    print(name, "name")
    print(agents, "agents")
    print(traits, "traits")
    print(strategy, "strategy")
    
    not_done = True
    while not_done:
        try:
            _ = init_and_run_experiment(
                        llm_key,
                        llm_dict[llm_key],
                        name,
                        name,
                        agents,
                        strategy,
                        traits,
                        data_type="test",
                    )
            not_done = False
        except Exception as E:
            print(E)
            time.sleep(0.5)
            not_done = True
```
